```python
import pandas as pd

# Import movies_metadata.csv
movies_metadata = pd.read_csv('movies_metadata.csv')

# Check out the first few rows of the keywords dataset
movies_metadata.head()
```

Each row gives us data (we will highlight some of the important ones here below):
- Adult. If the movie is for adults. 
- Collections it belongs to. This would be stuff like the Toy Story series and so on. 
- Budget. The amount budgeted for a movie, if it can be found. If data is not available, it is 0. 
- Generes. There's an id for the genre, along with the genre title. 
- id: id within the movies dataset
- imdb_id: id with IMDB
- Original Language: Language the movie was produced in
- Title: Name for the movie
- Overview: Synopsis of the movie
- Release Date: Date of release for the movie
- Revenue: How much revenue the movie made
- Runtime: How long the movie is
- Vote Count: Number of votes total cast for this movie
- Vote Average: Average score of votes

This data should prove to be useful in recommending movies. For example, a user may be predisposed to only want to see movies in a certain language, with a more recent release date, and high budget. We can take some of these factors in our rankings of movies and then recommend them to the user.
